```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
ims = []
ims.append(np.array(Image.open('baseline.png')))
ims.append(np.array(Image.open('min_loss.png')))
ims.append(np.array(Image.open('periodic.png')))
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,6))
for i in range(3):
    ax[i].imshow(ims[i])
    ax[i].axis('off')
fig.tight_layout()
```
